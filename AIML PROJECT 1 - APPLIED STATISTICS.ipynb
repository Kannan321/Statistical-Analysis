# <b>Statistical analysis<b>

 Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down? 
Write the null hypothesis and alternative hypothesis.
Test for significance and conclusion

<b>H0  = There is no significant difference between funds raised by operating and closed comapnies<br>
Ha = There is significant difference between funds raised by operating and closed comapnies</b>

There is no specific level of significance mention, so we consider the default 0.05 - 5%

```python
company_closed = company_df[company_df['OperatingState'] == 'Closed']['Funds_Million-$']
company_closed.reset_index(drop=True)
company_closed.describe()
```

```python
company_operating = company_df[company_df['OperatingState'] == 'Operating']['Funds_Million-$']
company_operating.reset_index(drop=True)
company_operating.describe()
```

```python
from statsmodels.stats.weightstats import ztest

alpha = 0.05 # level of significance
test_hypo , p_value = ztest(company_closed , company_operating)
```

```python
print("The P_value is : ",p_value)
```

```python
if p_value <= alpha:
    print("we reject the Null Hypothesis")
else:
    print("We fail to reject the Null Hypothesis")
```

The p_value is greater than the significance level , so we here fail to reject the null hypothesis.  so its clear that there is no significant difference in funds between the two classes.
